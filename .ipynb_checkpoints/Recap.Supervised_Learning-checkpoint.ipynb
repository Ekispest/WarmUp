## Валидация модели

```python
import numpy as np
import pandas as pd

from sklearn.datasets import make_classification
```

```python
Data, label = make_classification(n_samples=1000, 
                                  n_features=20, 
                                  n_informative=15, 
                                  n_redundant=3, 
                                  n_classes=2, 
                                  weights=[0.7, 0.3],
                                  flip_y=0.1)

# not necessary but convenient
Data = pd.DataFrame(Data)
label = pd.Series(label)

print(np.shape(Data)) 
label.value_counts()
```

### Обучаящая и тестовая выборки.  

```python
from sklearn.model_selection import train_test_split
```

```python
(trainData, 
 testData, 
 trainDependentVariables, 
 testDependentVariables) = train_test_split(Data,
                                            label,
                                            test_size=0.3)

print(trainData.shape,
     testData.shape,
     trainDependentVariables.shape,
     testDependentVariables.shape,
     trainDependentVariables.value_counts(normalize=True),
     testDependentVariables.value_counts(normalize=True),
     sep='\n')
```

### Подходы к валидации

** Shuffle method ** 

```python
from sklearn.model_selection import ShuffleSplit

cv = ShuffleSplit(n_splits=3, test_size=0.3, random_state=0)

cv.split(Data, label)
```

** Leave-P-out validation ** 

```python
from sklearn.model_selection import LeavePOut

lpo = LeavePOut(p=2)

lpo.split(Data, label)
```

** K-Fold validation ** 

```python
from sklearn.model_selection import KFold

kf = KFold(n_splits=5)

kf.split(Data, label)
```

** Cross-validation **

```python
from sklearn.dummy import DummyClassifier
from sklearn.model_selection import cross_val_score

clf  = DummyClassifier()

scores = cross_val_score(clf, Data, label, cv=5, scoring='f1_micro')
```

### Методы для настройки гиперпараметров модели

```python
from sklearn.model_selection import GridSearchCV

dm = DummyClassifier()
params = {'strategy': ['stratified', 'most_frequent', 'prior', 'uniform']}

clf = GridSearchCV(dm, params, cv=kf, scoring='f1_micro')
clf.fit(Data, label)

clf.cv_results_
```
